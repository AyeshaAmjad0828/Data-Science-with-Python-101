**W O R LD GDP PER CAPITA â€“ TASK BACKGROUND**

*Weâ€™ve collected data life_exp which contains the life expectancy for each country and
gdp_cap, which contains the GDP per capita (i.e. per person) for each country
expressed in US Dollars.
GDP stands for Gross Domestic Product. It basically represents the size of the
economy of a country. Divide this by the population and we get the GDP per capita*


a. Print the last item from both the list (GDP per Capita) gdp_cap, and the list (Life
Expectancy) life_exp.

b. Build a line chart, with gdp_cap and life_exp.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp_data = pd.read_csv('https://raw.githubusercontent.com/Masadn/PythonCourse/master/dataset/gapminder.csv')
```

```python
gdp_data
```

```python
life_exp=np.array(gdp_data['life_exp'])
```

```python
life_exp[-1]
```

```python
# Extracting columns
gdp_cap = gdp_data['gdp_cap']
life_exp = gdp_data['life_exp']
```

plt.plot(x,y)

```python
plt.plot(gdp_cap,life_exp)
plt.show()
```

```python
plt.scatter(gdp_cap,life_exp)
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
# Fitting a logarithmic trend line
coefficients = np.polyfit(np.log(gdp_cap + 1), life_exp, 1)
log_fit = np.poly1d(coefficients)

# Generating points for the trend line
gdp_cap_sorted = np.sort(gdp_cap)
life_exp_fit = log_fit(np.log(gdp_cap_sorted + 1))

# Plotting the trend line
plt.plot(gdp_cap_sorted, life_exp_fit, color='red', label='Logarithmic Fit')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.title('Life Expectancy vs GDP per Capita with Logarithmic Fit')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

#male = 0
#female = 1

age = [1, 2, 3, 4, 5]
height = [2, 3, 4, 5, 6]
weight = [20, 400, 80, 200, 800]
gender = [0, 0, 1, 0, 1]

plt.scatter(age, height, s=weight, c=gender, alpha=0.5, marker='o', edgecolors='w', linewidths=2)
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Scatter Plot of Age, Height, Weight, and Gender')
plt.show()
```

#Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

Let's discuss some plots that allow us to visualize the data set. These plots are:

* Scatter Plot
* Line Plot
* Bar Graph
* Histogram
* Heatmap

```python
import seaborn as sn
```

```python
car.head(10)
```

### Selling Price by Fuel Type

```python
ascending_sp = car.groupby('fuel')['selling_price'].mean().sort_values()
```

```python
sn.barplot(x='fuel', y='selling_price', data=car, order=ascending_sp.index)
```

```python
dscending_sp = car.groupby('fuel')['selling_price'].mean().sort_values(ascending=False)
```

```python
sn.barplot(x='fuel', y='selling_price', data=car, order=dscending_sp.index)
```

### Fuel Type Distribution

```python
count_of_fuel_type = car['fuel'].value_counts()
count_of_fuel_type
```

```python
plt.figure(figsize=(10,6))
count_of_fuel_type.plot(kind='bar', color='skyblue')
plt.title('Fuel Type Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Count of Cars')
plt.show()
```

### Seller Type Distribution

```python
count_of_seller_type= car['seller_type'].value_counts()
count_of_seller_type
```

```python
plt.figure(figsize=(10,6))
count_of_seller_type.plot(kind='bar')
plt.title('Seller Type Distribution')
plt.xlabel('Seller Type')
plt.ylabel('Count of Cars')
plt.show()
```

### Transmission Type Distribution
