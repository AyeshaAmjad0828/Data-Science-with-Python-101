
###Task# 4

a. Load advertising data.

    i. TV vs Sales

    ii. Social Media vs Sales

    iii. Newspaper vs Sales

c. Which advertising expenditures is most correlated with sales?

d. Which advertising expenditure is least correlated with sales?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
advertising = pd.read_csv('/content/sample_data/Advertising.csv')
advertising.head()
```

```python
advertising.drop(columns = 'Unnamed: 4', inplace = True)
advertising.head()
```

```python
plt.scatter(advertising['TV(millionUSD)'], advertising['sales(millionUSD)'])
plt.title('TV expenditure vs Sales')
plt.xlabel('TV expenditure')
plt.ylabel('Sales')
plt.show()
```

In TV vs Sales scatter plot, we observe a positively strong correlation.

```python
plt.scatter(advertising['SocialMedia(millionUSD)'], advertising['sales(millionUSD)'])
plt.title('Social Media expenditure vs Sales')
plt.xlabel('Social Media expenditure')
plt.ylabel('Sales')
plt.show()
```

```python
plt.scatter(advertising['newspaper(millionUSD)'], advertising['sales(millionUSD)'])
plt.title('newspaper expenditure vs Sales')
plt.xlabel('newspaper expenditure')
plt.ylabel('Sales')
plt.show()
```

```python
# Create scatter plots
plt.figure(figsize=(15, 5))

# TV vs Sales
plt.subplot(1, 3, 1)
plt.scatter(advertising['TV(millionUSD)'], advertising['sales(millionUSD)'])
plt.title('TV vs Sales')
plt.xlabel('TV (million USD)')
plt.ylabel('Sales (million USD)')

# Social Media vs Sales
plt.subplot(1, 3, 2)
plt.scatter(advertising['SocialMedia(millionUSD)'], advertising['sales(millionUSD)'])
plt.title('Social Media vs Sales')
plt.xlabel('Social Media (million USD)')
plt.ylabel('Sales (million USD)')

# Newspaper vs Sales
plt.subplot(1, 3, 3)
plt.scatter(advertising['newspaper(millionUSD)'], advertising['sales(millionUSD)'])
plt.title('Newspaper vs Sales')
plt.xlabel('Newspaper (million USD)')
plt.ylabel('Sales (million USD)')

# Show plot
plt.tight_layout()
plt.show()
```

## Data Distributions

Data distribution is a spread of data and the data points.By examining the distribution, we can gain insights into the characteristics and patterns of the data, which can be useful in making informed decisions and predictions such as mean, median, variance, standard deviation, skewness and kurtosis. There are various types of data distributions, each with its own unique properties and implications. Understanding these distributions is a fundamental aspect of data analysis and can help us make more accurate and meaningful interpretations of the data.

###Task# 6

Create histograms on all float columns in Advertising data.

```python
##MATPLOTLIB

plt.hist(advertising['sales(millionUSD)'], bins = 20, color = 'skyblue', edgecolor = 'black')
plt.title('Sales Data Distribution')
plt.ylabel('Sales (million USD)')
plt.show()
```

```python
##SEABORN

sns.histplot(advertising['sales(millionUSD)'], bins = 20, color = 'skyblue', edgecolor = 'black', kde=True)
plt.title('Sales Data Distribution')
plt.ylabel('Sales (million USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
##MATPLOTLIB

# Create scatter plots
plt.figure(figsize=(15, 5))

# TV vs Sales
plt.subplot(1, 3, 1)
plt.hist(advertising['TV(millionUSD)'], edgecolor = 'black')
plt.title('TV')
plt.ylabel('TV (million USD)')


# Social Media vs Sales
plt.subplot(1, 3, 2)
plt.hist(advertising['SocialMedia(millionUSD)'], edgecolor = 'black')
plt.title('Social Media')
plt.ylabel('Social Media (million USD)')


# Newspaper vs Sales
plt.subplot(1, 3, 3)
plt.hist(advertising['newspaper(millionUSD)'], edgecolor = 'black')
plt.title('Newspaper')
plt.ylabel('Newspaper (million USD)')


# Show plot
plt.tight_layout()
plt.show()
```

```python
##SEABORN

# Create scatter plots
plt.figure(figsize=(15, 5))

# TV vs Sales
plt.subplot(1, 3, 1)
sns.histplot(advertising['TV(millionUSD)'], edgecolor = 'black', kde=True)
plt.title('TV')
plt.xlabel('TV (million USD)')
plt.ylabel('Frequency')


# Social Media vs Sales
plt.subplot(1, 3, 2)
sns.histplot(advertising['SocialMedia(millionUSD)'], edgecolor = 'black', kde=True)
plt.title('Social Media')
plt.xlabel('Social Media (million USD)')
plt.ylabel('Frequency')

# Newspaper vs Sales
plt.subplot(1, 3, 3)
sns.histplot(advertising['newspaper(millionUSD)'], edgecolor = 'black', kde=True)
plt.title('Newspaper')
plt.xlabel('Newspaper (million USD)')
plt.ylabel('Frequency')

# Show plot
plt.tight_layout()
plt.show()
```
